#setting the parameters for the two csv's 
names_vntr = ['header', 'tag', 'trid', 'reedid', 'reedstart', 'reedstop','trid2','chr', 'firstindex', 'lastindex', 'subheader']
names_marzie = ['tag', 'chr', 'start', 'end', 'fragments', 'num1', 'num2', 'calulation']
dtypes_marzie  = {'tag': 'str', 'chr':'str', 'start':'int', 'end':'int', 'fragments':'int', 'num1':'int', 'num2':'int', 'calulation':'float'}
dtypes_vntr = {'header':'str', 'tag':'str', 'trid':'int','reedid':'int', 'reedstart':'int', 'reedstop':'int','trid2':'int','chr':'str', 'firstindex':'int', 'lastindex':'int', 'subheader':'str'}

#loading the two csv's
marzie_10x = pd.read_csv('barcodes_mapping_witho_chr.csv',names = names_marzie, dtype = dtypes_marzie)
vntr = pd.read_csv('sorted_without_chr_vntr.csv',names = names_vntr, dtype = dtypes_vntr)


#can read a line at a time
#test1 = pd.read_csv('untitled.csv', names = ['name','num'], dtype = {'name': 'str', 'num':'int'})

#sort by tags and chr 
sorted_vntr = vntr.sort_values(by= ['tag','chr'])
sorted_marzie = marzie_10x.sort_values(by = ['tag', 'chr'])


#setting another dataframe for the output


#opening the necessary files 
intersected_dynamic = open('intersected_dynamic_test2.csv', 'a') 
intersected_failed = open('intersected_failed_dynamic_test2.csv', 'a')
no_tag_found = open('no_tag_found_intersect_dynamic_test2.csv', 'a')

intersected_dynamic.write('tag,chr,marzie_start,marzie_stop,reedstart,reedstop,trid,reed_header,vntr_chr' + "\n")
intersected_failed.write('header,tag,trid,reedid,reedstart,reedstop,vntr_chr,firstindex,lastindex' + "\n")
no_tag_found.write('header,tag,trid,reedid,reedstart,reedstop,vntr_chr,firstindex,lastindex' + "\n")


#function 

for x in range(len(sorted_vntr.header)): # for the number of reeds in vntr_file
    tag1 = sorted_vntr.tag[x] # get their tags
    line_vntr = sorted_vntr.iloc[x] # get that associated line
    line1_marzie = sorted_marzie.loc[sorted_marzie['tag'] == str(tag1)] # check if marzie's file has the vntr tag

    if line1_marzie.empty == False: # if it does
        matching_chromosomes = line1_marzie
        reedstart_vntr = sorted_vntr.iloc[x].reedstart
        reedstop_vntr = sorted_vntr.iloc[x].reedstop
        num_of_tags = len(matching_chromosomes.tag) #number of tags (if there are 5 tags this will be 5)
        mismatched_marzie_reeds= 0
        for y in range(num_of_tags): # need to iterate over all the matching tags that show up "[0,1,2,3,4]"
            reed_start_marzie = matching_chromosomes.iloc[y].start
            reed_stop_marzie = matching_chromosomes.iloc[y].end
            matching_line_marzie= line1_marzie.iloc[y]
            if matching_line_marzie.chr == line_vntr.chr and int(reedstart_vntr) in range(int(reed_start_marzie),int(reed_stop_marzie)): # check to see if the vntr range is in the marzie range and have the same chr

                intersected_dynamic.write(str(matching_line_marzie.tag)+','+str(matching_line_marzie.chr) + ',' + str(reed_start_marzie)+ ',' +str(reed_stop_marzie) + ',' +str(line_vntr.reedstart) + ',' +str(line_vntr.reedstop) + ',' +str(line_vntr.trid)+ ',' +str(line_vntr.header)+ ',' +str(line_vntr.chr)+"\n")
                # if it is in the range and has the same chromosome then add to the intersected data frame
            else:
                mismatched_marzie_reeds = mismatched_marzie_reeds + 1 # will count how many mismatched_tags have been processed so far
                if mismatched_marzie_reeds == num_of_tags: # if you cycle through all the tags

                    intersected_failed.write(str(line_vntr.header)+','+str(line_vntr.tag) + ',' + str(line_vntr.trid)+ ',' +str(line_vntr.reedid) + ',' +str(line_vntr.reedstart) + ',' +str(line_vntr.reedstop) + ',' +str(line_vntr.chr)+ ',' +str(line_vntr.firstindex)+ ',' +str(line_vntr.lastindex)+"\n")

                #else if it has cycled through all the matched tags and has not found a match add that vntr reed to a failed dataframe
    else: # it it didnt even find any tags add it to the failed data frame

        no_tag_found.write(str(line_vntr.header)+','+str(line_vntr.tag) + ',' + str(line_vntr.trid)+ ',' +str(line_vntr.reedid) + ',' +str(line_vntr.reedstart) + ',' +str(line_vntr.reedstop) + ',' +str(line_vntr.chr)+ ',' +str(line_vntr.firstindex)+ ',' +str(line_vntr.lastindex)+"\n")

        
intersected_dynamic.close()
intersected_failed.close()
no_tag_found.close()
